```python
import requests
from bs4 import BeautifulSoup #Procesammiento de página y obtener contenido
```

```python
url = 'https://pcel.com//' 
```

```python
#Peticion de URL
pcel = requests.get(url) #get, post, put delete
```

```python
pcel.status_code #un status code de 200 es una respuesta exitosa
```

```python
print(pcel.text)
```

```python
print(pcel.headers)
```

```python
#Preprocesamos (Transformación)
s = BeautifulSoup(pcel.text,'lxml')
```

```python
#imprime de manera legible lo que trajo
print(s.prettify())
```

```python
## Conocemos las secciones o links del menu
secciones = s.find_all("a",{"class":"pcel-dropdown-item"})
secciones
```

```python
#Obtenemos cada elemento de las secciones.
#<a href="url o link">Texto</a>
link_sections = [item.get('href') for item in secciones]
title_sections = [item.get_text() for item in secciones]
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.latarde.com.mx/'
```

```python
##Peticion o request a la pagina##
latarde = requests.get(url)
```

```python
latarde.status_code #un status de 200 es una respuesta exitosa
```

```python
print(latarde.text)
```

```python
print(latarde.headers)
```

```python
##Preprocesamiento(transformacion)
s = BeautifulSoup(latarde.text,'lxml')
s
```

```python
print(s.prettify())
```

```python
secciones = s.find_all("meta",{"content":"La Tarde"})
secciones
```

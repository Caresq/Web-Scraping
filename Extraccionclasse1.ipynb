##Construcción de un Web Scrapper##


```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.pagina12.com.ar'
```

```python
##Peticion o request a la pagina##
p12 = requests.get(url)
```

```python
p12.status_code #un status de 200 es una respuesta exitosa
```

```python
print(p12.text)
```

```python
print(p12.headers)
```

```python
##Preprocesamiento(transformacion)
s = BeautifulSoup(p12.text,'lxml')
s
```

```python
print(s.prettify())
```

```python
secciones = s.find_all("a",{"class":"p12-dropdown-item"})
secciones
```
